```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df=pd.read_csv('canada_per_capita_income.csv')
df.head()
```

```python
#Plot a Scatter plot
%matplotlib inline
plt.scatter(df['year'],df['per capita income (US$)'],color='Red',marker='+')
plt.ylabel('per capita income (US$)')
plt.xlabel('Year')
```

```python
#Create linear Regression object
reg=linear_model.LinearRegression()
reg.fit(df[['year']],df['per capita income (US$)'])
```

```python
reg.predict([[2020]])
```

```python
#Check Value of m in y=mx+b
reg.coef_
```

```python
#Check value of intercept
reg.intercept_
```

```python
d=pd.read_csv('Year.csv')
d.head()
```

```python
p=reg.predict(d)
```

```python
d['per capita income (US$)']=p
```

```python
d.to_csv("Year.csv",index=False)
```

```python
#Check the Linear regression line on Scatterplot
%matplotlib inline
plt.scatter(df['year'],df['per capita income (US$)'],color='red',marker='+')
plt.plot(df['year'],reg.predict(df[['year']]),color='blue')
```
